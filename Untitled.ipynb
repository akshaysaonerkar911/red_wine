```python
sns.boxplot(df["residual sugar"])  # given column outliers
```

```python
# IQR method to remove outliers

q3 = np.percentile(df["residual sugar"],q = 75)
q1 = np.percentile(df["residual sugar"],q = 25)
iqr = q3 -q1
print(iqr)
upper_tail = q3 + iqr * 1.5
lower_tail = q1 - iqr * 1.5
print("upper_tail>>",upper_tail)
print("lower_tail>>",lower_tail)
```

```python
df.loc[(df["residual sugar"] > upper_tail)]
x = df.loc[(df["residual sugar"] < upper_tail)]["residual sugar"].mean()
df.loc[(df["residual sugar"] > upper_tail,"residual sugar")] = x
```

```python
df.loc[(df["residual sugar"] > upper_tail)] # outliers is removed
```

```python
q3 = np.percentile(df["free sulfur dioxide"],q = 75)
q1 = np.percentile(df["free sulfur dioxide"],q = 25)
iqr = q3 -q1
print(iqr)
upper_tail = q3 + iqr * 1.5
lower_tail = q1 - iqr * 1.5
print("upper_tail>>",upper_tail)
print("lower_tail>>",lower_tail)
df.loc[(df["free sulfur dioxide"] > upper_tail)]
x = df.loc[(df["free sulfur dioxide"] < upper_tail)]["free sulfur dioxide"].mean()
df.loc[(df["free sulfur dioxide"] > upper_tail,"free sulfur dioxide")] = x
```

```python
df.loc[(df["free sulfur dioxide"] > upper_tail)]
```

```python
q3 = np.percentile(df["total sulfur dioxide"],q = 75)
q1 = np.percentile(df["total sulfur dioxide"],q = 25)
iqr = q3 -q1
print(iqr)
upper_tail = q3 + iqr * 1.5
lower_tail = q1 - iqr * 1.5
print("upper_tail>>",upper_tail)
print("lower_tail>>",lower_tail)
df.loc[(df["total sulfur dioxide"] > upper_tail)]
x = df.loc[(df["free sulfur dioxide"] < upper_tail)]["total sulfur dioxide"].mean()
df.loc[(df["total sulfur dioxide"] > upper_tail,"total sulfur dioxide")] = x
```

```python
# creating derived features

df["volatile_mean"] = df["volatile acidity"].mean()
df["citric_mean"] = df["citric acid"].mean()
df["residual_mean"] = df["residual sugar"].mean()
df["chloride_mean"] = df["chlorides"].median()
df["free_mean"] = df["free sulfur dioxide"].median()
df["total_mean"] = df["total sulfur dioxide"].median()
```

```python
# distribute independant and dependat variables
x = df.drop("quality",axis= 1)
y = df["quality"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=65)  # splitting data into training and testing for evaluation purpose
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Model training
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
# evaluation of training data
y_pred_train = log_model.predict(x_train)
y_pred_train
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("conf>>\n",conf)
clas = classification_report(y_train,y_pred_train)
print("classificatio>>",clas)
```

```python
# testing evaluation matrix
# evaluation of training data
y_pred_test = log_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("conf>>\n",conf)
clas = classification_report(y_test,y_pred_test)
print("classificatio>>",clas)
```

```python
# logistic regression not perform well got less accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(max_depth=9, min_samples_leaf=10)
dt_model.fit(x_train,y_train)
```

```python
# evaluation of training data
y_pred_train = dt_model.predict(x_train)
y_pred_train
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("conf>>\n",conf)
clas = classification_report(y_train,y_pred_train)
print("classificatio>>",clas)
```

```python
# testing evaluation matrix
# evaluation of training data
y_pred_test = dt_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("conf>>\n",conf)
clas = classification_report(y_test,y_pred_test)
print("classificatio>>",clas)
```

```python
# model is overfitting so i will tuned the model
```

```python
dt_model = DecisionTreeClassifier()
hyp = {"criterion":["gini","entropy"],
       "min_samples_split":np.arange(2,20),
       'min_samples_leaf':np.arange(2,15),
       "max_depth":np.arange(2,10)}
rscv_model = RandomizedSearchCV(dt_model,hyp,cv = 5)
rscv_model.fit(x_train,y_train)
rscv_model.best_estimator_
```

```python
# on Desicion tree also not given good accuracy
```

```python
# Randomforest model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(criterion='entropy', max_depth=8, max_features='log2',
                       min_samples_leaf=6, min_samples_split=5, n_estimators=34,
                       oob_score=True)
rf_model.fit(x_train,y_train)
```

```python
# evaluation of training data
y_pred_train = rf_model.predict(x_train)
y_pred_train
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("conf>>\n",conf)
clas = classification_report(y_train,y_pred_train)
print("classificatio>>",clas)
```

```python
# testing evaluation matrix
# evaluation of training data
y_pred_test = rf_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("conf>>\n",conf)
clas = classification_report(y_test,y_pred_test)
print("classificatio>>",clas)
```

```python
# model is overfitting so i tuned the data
```

```python
rf_model = RandomForestClassifier()
hyp = {"n_estimators":np.arange(2,100),
    "criterion":["gini","entropy"],
       "min_samples_split":np.arange(2,20),
       'min_samples_leaf':np.arange(2,15),
       "max_depth":np.arange(2,10),"max_features":["log2","auto"],
      "oob_score":[True],'bootstrap':[True]}
rscv_model = RandomizedSearchCV(rf_model,hyp,cv = 5)
rscv_model.fit(x_train,y_train)
rscv_model.best_estimator_
```
